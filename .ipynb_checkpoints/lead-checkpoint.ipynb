```python
X = data.iloc[:, :-1]
X.head()
```

```python
X.describe(include=np.object).columns
```

```python
X = pd.get_dummies(X, columns=['InitialContact', 'Title', 'LeadSource', 'ExistingCustomer', 'Region',
       'CampaignMix'], dtype='int', drop_first=True)
```

```python
X.head()
```

```python
X.shape
```

```python
X.describe()
```

```python
Y = data.iloc[:, -1]
Y.head()
```

```python
#Encoding dependent variable
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
print('X_train shape: ', X_train.shape )
print('X_test shape: ', X_test.shape )
print('Y_train shape: ', Y_train.shape )
print('Y_test shape: ', Y_test.shape )
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

Functions to get the base scores for each model

```python
def get_base_prediction(model_dict):
    scores = {}
    for name, model in model_dict.items():
        model = Pipeline(steps=[
                        ('model', model)
        ])
        
        model.fit(X_train, Y_train)
        scores[name] = f'{model.score(X_test, Y_test) * 100}%'
        
    return scores

get_base_prediction({'random_forest': RandomForestClassifier(random_state=0),
                    'logistic_regression': LogisticRegression(random_state=0),
                    'decision_tree': DecisionTreeClassifier(random_state=0)})
```
